```python

textfile = "input.txt"

position = 0
depth = 0
aim = 0
for l in open(textfile):
    l = l.split()
    amount = int(l[1])
    if l[0] == "forward":
        position += amount
        depth += aim * amount
    elif l[0] == "up":
        depth -= amount
    elif l[0] == "down":
        depth += amount
        aim += amount
```
